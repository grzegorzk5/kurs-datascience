Masz string CSV: "Alice,28,Data Scientist,75000" . Użyj split(',') aby wyciągnąć:
Imię
Wiek (jako int)
Zawód
Pensję (jako int)
Wyświetl te informacje w formacie: "Alice ma 28 lat, pracuje jako Data Scientist i
zarabia $75000"

```python
csv_string = "Alice,28,Data Scientist,75000"

# Sposób 1
# imie, wiek, zawod, pensja = csv_string.split(',')
# wiek = int(wiek)
# pensja = int(pensja)

# Sposób 2
imie, wiek, zawod, pensja = [int(string) if string.isdecimal() else string for string in csv_string.split(',')]

print(f"Typ zmiennej imie: {type(imie)}")
print(f"Typ zmiennej wiek: {type(wiek)}")
print(f"Typ zmiennej zawod: {type(zawod)}")
print(f"Typ zmiennej pensja: {type(pensja)}")
print()

print(f"{imie} ma {wiek} lat, pracuje jako {zawod} i zarabia ${pensja}")
```
