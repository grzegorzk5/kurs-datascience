Zacznij od pustej listy results = [] . Dodaj do niej 5 wartości accuracy: 0.75, 0.82, 0.79,
0.85, 0.88. Następnie:
Wyświetl długość listy
Wyświetl średnią wartość
Wyświetl maksymalną wartość

```python
results = []
accuracy = [0.75, 0.82, 0.79, 0.85, 0.88]

# Sposób 1
for value in accuracy:
    results.append(value)

# Sposób 2
# results.extend(accuracy)

print(f"Długość listy results: {len(results)}")
print(f"Średnia wartość w results: {sum(results)/len(results):.2f}")
print(f"Maksymalna wartość w results: {max(results)}")
```
