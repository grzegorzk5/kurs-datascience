Masz listę wyników: [0.78, 0.92, 0.85, 0.88, 0.81] . Posortuj ją malejąco i wyświetl 3
najlepsze wyniki.

```python
scores = [0.78, 0.92, 0.85, 0.88, 0.81]

# Sposób 1
# scores.sort(reverse=True)

# Sposób 2
scores_sorted = sorted(scores, reverse=True)

print(f"Wyniki posortowane malejąco: {scores_sorted}")
print(f"Trzy najlepsze wyniki: {scores_sorted[:3]}")
```
