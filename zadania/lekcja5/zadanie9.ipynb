Masz listę pomiarów: [23, 24, 22, 150, 23, 25, -10, 24, 23] . Outlierem jest wartość
poza zakresem [20, 30]. Stwórz nową listę bez outlierów używając list comprehension.

```python
measurements = [23, 24, 22, 150, 23, 25, -10, 24, 23]
measurements_without_outliers = [measurement for measurement in measurements if measurement >= 20 and measurement <= 30]
outliers = [measurement for measurement in measurements if measurement not in measurements_without_outliers]


print(f"Lista pomiarów z outlierami: {measurements}")
print(f"Lista pomiarów bez outlierów: {measurements_without_outliers}")

print(f"Outliers: {outliers}")
```
