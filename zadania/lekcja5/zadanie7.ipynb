Masz listę predykcji: ['spam', 'ham', 'spam', 'spam', 'ham', 'ham', 'spam'] .
Policz ile jest 'spam' i ile 'ham' używając metody count() .

```python
predictions = ['spam', 'ham', 'spam', 'spam', 'ham', 'ham', 'spam']
spam_count = predictions.count('spam')
ham_count = predictions.count('ham')

print(f"Ilość spam = {spam_count}, Ilość ham = {ham_count}")
```
