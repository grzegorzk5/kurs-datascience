Stwórz listę zawierającą pomiary temperatury z 7 dni: 
[22.5, 23.0, 21.5, 24.0, 23.5, 22.0, 23.0] . 
Wyświetl:
Temperaturę z pierwszego dnia
Temperaturę z ostatniego dnia
Temperatury z trzech środkowych dni (slicing)

```python
pomiary_temperatury = [22.5, 23.0, 21.5, 24.0, 23.5, 22.0, 23.0]

# print(f"Pomiary temperatury: {list(map(lambda val: "{0} °C".format(val), pomiary_temperatury))}")
# print(f"Pomiary temperatury: {list(map(lambda val: f'{val} °C', pomiary_temperatury))} ")
print(f"Pomiary temperatury: {[f'{val} °C' for val in pomiary_temperatury]}")
print(f"Temperatura z pierwszego dnia: {pomiary_temperatury[0]} °C")
print(f"Temperatura z ostatniego dnia: {pomiary_temperatury[-1]} °C")
print(f"Temperatura z trzech środkowych dni: {[f'{val} °C' for val in pomiary_temperatury[2:5]]}")
```
