Masz listę kategorii produktów z różnymi formatami:

categories = ["  Electronics  ", "ELECTRONICS", "electronics", "  Books", "books  ", "BOOKS"]

Znormalizuj je (strip + lower) i znajdź unikalne kategorie.


```python
categories = ["  Electronics  ", "ELECTRONICS", "electronics", "  Books", "books  ", "BOOKS"]

print(f"Kategorie produktów bez normalizacji: {categories}")

categories_normalized = [category.strip().lower() for category in categories]
print(f"Znormalizowane kategorie produktów: {categories_normalized}")

# Sposób 1
# unique_categories = []

# for category in categories_normalized:
#     if category not in unique_categories:
#         unique_categories.append(category)

# Sposób 2
unique_categories = list(set(categories_normalized))
print(f"Unikalne kategorie produktów: {unique_categories}")
```
