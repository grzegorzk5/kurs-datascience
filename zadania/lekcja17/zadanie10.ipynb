Utwórz wykres ceny akcji w czasie z:
- Główną linią ceny
- Średnią kroczącą 7-dniową (MA7)
- Średnią kroczącą 21-dniową (MA21)
- Zacienionym obszarem między MA7 a MA21

````python
import numpy as np
import pandas as pd
import plotly.graph_objects as go

# Przygotowanie danych
np.random.seed(42)
dates = pd.date_range('2025-01-01', periods=100)
price = 100 + np.cumsum(np.random.randn(100))
df = pd.DataFrame({'date': dates, 'price': price})

# Obliczanie średnich kroczących
df['MA7'] = df['price'].rolling(window=7).mean()
df['MA21'] = df['price'].rolling(window=21).mean()

# Tworzenie wykresu
fig = go.Figure()

# Główna linia ceny
fig.add_trace(go.Scatter(x=df['date'], y=df['price'], name='Cena', 
                         line=dict(color='black', width=1), opacity=0.4))

# MA21 (dolna lub górna granica obszaru)
fig.add_trace(go.Scatter(x=df['date'], y=df['MA21'], name='MA21', 
                         line=dict(color='red', width=1.5)))

# MA7 z wypełnieniem do MA21
fig.add_trace(go.Scatter(x=df['date'], y=df['MA7'], name='MA7', 
                         line=dict(color='blue', width=1.5),
                         fill='tonexty', fillcolor='rgba(128, 128, 128, 0.2)'))

# Formatowanie
fig.update_layout(
    title='Wykres ceny akcji ze średnimi kroczącymi',
    xaxis_title='Data',
    yaxis_title='Cena',
    hovermode='x unified',
    template='plotly_white'
)

fig.show()
````
