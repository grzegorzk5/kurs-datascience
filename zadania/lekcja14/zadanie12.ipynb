##### Rozwiązanie dla jednej bazy danych zarządzanej przez Alembic:
- stworzenie struktury plików migracji

uv run alembic init

- Konfiguracja:

 \- ustawienie URL do swojej bazy danych w pliku alembic.ini

 \- konfiguracja metadanych, aby alembic widział moje modele
alembic/env.py

- dodanie modelu Article (id, title, content) do models.py
    
class Article(Base):
    __tablename__ = 'articles'

    id = Column(Integer, primary_key=True, autoincrement=True)
    title = Column(String(255), nullable=False)
    content = Column(Text)
    

- Wygenerowanie pierwszego pliku migracji na podstawie modelu

uv run alembic revision --autogenerate -m "Create articles table"

- Zastosowanie migracji
uv run alembic upgrade head

Po wykonaniu upgrade baza będzie zarządzana przez system migracji.Stworzona zostanie tabela alembic_version oraz tabela
articles

- dodanie kolumny published_date do modelu Article (models.py)

```
# ...
published_date = Column(DateTime, nullable=True)
# ...
```

- Wygenerowanie kolejnej migracji. Alembic wykryje różnicę między modelem, a bazą

alembic revision --autogenerate -m "Add published_date to articles"

- zastosowanie migracji - dodanie nowej kolumny bez kasowania danych

uv run alembic upgrade head
