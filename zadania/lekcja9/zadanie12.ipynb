Stwórz klasę bazową Figura z metodą pole() .

Stwórz klasy potomne:
- Prostokat(szerokosc, wysokosc)
- Kolo(promien)

Każda klasa implementuje pole() .

Stwórz listę figur i oblicz łączne pole.

```python
class Figura:
    def __init__(self, nazwa: str):
        self.nazwa: str = nazwa

    def pole(self) -> float:
        raise NotImplementedError("Metoda pole() musi być zaimplementowana!")

class Prostokat(Figura):
    def __init__(self, szerokosc: float, wysokosc: float):
        super().__init__("Prostokąt")
        self.szerokosc: float = szerokosc
        self.wysokosc: float = wysokosc

    def pole(self) -> float:
        return self.szerokosc * self.wysokosc
    
class Kolo(Figura):
    def __init__(self, promien: float):
        super().__init__("Koło")
        self.promien: float = promien
    
    def pole(self) -> float:
        return 3.14 * (self.promien ** 2)

szerokosc_prostokata: float = 4
wysokosc_prostokata: float = 5
promien_kola:float = 6

figury: list[Figura] = [
    Prostokat(szerokosc_prostokata, wysokosc_prostokata),
    Kolo(promien_kola)
]

laczne_pole_figur: float = 0
for figura in figury:
    laczne_pole_figur += figura.pole()

print(f"Łączne pole figur (prostokąta i koła) wynosi: {laczne_pole_figur:.2f}")
```
