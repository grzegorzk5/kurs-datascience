Stwórz klasę Pracownik z atrybutami: imie , stanowisko , pensja .

Metody:
- podwyzka(procent) – zwiększa pensję o procent%
- __str__() – reprezentacja tekstowa

Stwórz 3 pracowników i daj im podwyżki.

```python
class Pracownik:
    def __init__(self, imie: str, stanowisko: str, pensja: float):
        self.imie = imie
        self.stanowisko = stanowisko
        self.pensja = pensja

    def podwyzka(self, procent: float):
        self.pensja = self.pensja * (1 + (procent/100))

    def __str__(self) -> str:
        return f"Imię: {self.imie}, Stanowisko: {self.stanowisko}, Pensja: {self.pensja:.2f}"

pracownik1 = Pracownik("Jan", "Data Scientist", 6500)
pracownik2 = Pracownik("Michał", "Machine Learning Engineer", 9000)
pracownik3 = Pracownik("Jerzy", "Software Engineer", 7500)

pracownik1.podwyzka(9)
print(pracownik1)

pracownik2.podwyzka(12)
print(pracownik2)

pracownik3.podwyzka(7)
print(pracownik3)
```
