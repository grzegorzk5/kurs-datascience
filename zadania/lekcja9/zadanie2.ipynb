Stwórz klasę Kalkulator z metodami: dodaj(a, b) , odejmij(a, b) , pomnoz(a, b) , podziel(a, b) .

Stwórz obiekt i wykonaj przykładowe obliczenia.

```python
class Kalkulator:
    def dodaj(self, a:float, b:float) -> float:
        return a + b
    
    def odejmij(self, a:float, b:float) -> float:
        return a - b
    
    def pomnoz(self, a:float, b:float) -> float:
        return a * b
    
    def podziel(self, a:float, b:float) -> float:
        return a / b
    

kalkulator = Kalkulator()

print(f"4 + 5 = {kalkulator.dodaj(4, 5)}")
print(f"2 - 5 = {kalkulator.odejmij(2, 5)}")
print(f"4 * 5 = {kalkulator.pomnoz(4, 5)}")
print(f"12 / 5 = {kalkulator.podziel(12, 5)}")
```
