Stwórz 4 komponenty:
- FiltrowanieDodatnich – usuwa liczby ujemne
- Zaokraglanie – zaokrągla do całkowitych
- Skalowanie – mnoży przez stały współczynnik
- Sumowanie – zwraca sumę listy

Stwórz klasę Pipeline i przetestuj na danych: [1.7, -3.2, 5.8, -1.1, 8.3, 2.4]

```python
class FiltowanieDodatnich:
    def __init__(self):
        self.nazwa = "Filtrowanie"

    def wykonaj(self, dane):
        print(f"  [{self.nazwa}] Usuwam wartości ujemne...")
        return [wartosc for wartosc in dane if wartosc >= 0]

class Zaokraglanie:
    def __init__(self):
        self.nazwa = "Zaokrąglanie"

    def wykonaj(self, dane):
        print(f"  [{self.nazwa}] Zaokrąglam do liczb całkowitych...")
        return [round(wartosc) for wartosc in dane]

class Skalowanie:
    def __init__(self, wspolczynnik):
        self.nazwa = "Skalowanie"
        self.wspolczynnik = wspolczynnik

    def wykonaj(self, dane):
        print(f"  [{self.nazwa}] Skaluję dane...")
        return [wartosc * self.wspolczynnik for wartosc in dane]

class Sumowanie:
    def __init__(self):
        self.nazwa = "Sumowanie"

    def wykonaj(self, dane):
        print(f"  [{self.nazwa}] Sumuję liczby...")
        return sum(dane)

class Pipeline:
    def __init__(self, nazwa: str):
        self.nazwa = nazwa
        self.kroki = []
    
    def dodaj_krok(self, krok):
        self.kroki.append(krok)

    def uruchom_krok(self, dane):
        print(f"=== Pipeline: {self.nazwa} ===")
        wynik = dane

        for krok in self.kroki:
            wynik = krok.wykonaj(wynik)
        
        return wynik


dane = [1.7, -3.2, 5.8, -1.1, 8.3, 2.4]

pipeline = Pipeline("Przetwarzanie danych")
pipeline.dodaj_krok(FiltowanieDodatnich())
pipeline.dodaj_krok(Zaokraglanie())
pipeline.dodaj_krok(Skalowanie(10))
pipeline.dodaj_krok(Sumowanie())

print(f"Dane: {dane}\n")

wynik = pipeline.uruchom_krok(dane)

print(f"\nWynik po przetworzeniu: {wynik}")
```
