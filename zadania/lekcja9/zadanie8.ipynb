Stwórz klasę Zadanie z atrybutami: opis , status (domyślnie "do zrobienia").

Metody:
- oznacz_jako_zrobione() – zmienia status na "zrobione"
- __str__() – wyświetla opis i status

Stwórz listę 5 zadań i oznacz część jako zrobione.

```python
class Zadanie:
    def __init__(self, opis: str, status: str ="do zrobienia"):
        self.opis = opis
        self.status = status

    def oznacz_jako_zrobione(self):
        self.status = "zrobione"
    
    def __str__(self) -> str:
        return f"Opis: {self.opis}, Status: {self.status}"
    
lista_zadan = [
    Zadanie("Zakończyć prezentację projektu dla klienta (do 11:00)"),
    Zadanie("Odpisać na zaległe maile", "do zrobienia"),
    Zadanie("Zapłać rachunki (prąd/woda/internet)"),
    Zadanie("Uporządkować miejsce pracy", "do zrobienia"),
    Zadanie("Optymalizacja zapytań do bazy danych", "do zrobienia")
]

lista_zadan[0].oznacz_jako_zrobione()
lista_zadan[1].oznacz_jako_zrobione()
lista_zadan[3].oznacz_jako_zrobione()

print("Lista zadań do zrobienia")

for nr_zadania, zadanie in enumerate(lista_zadan):
    print(f"Zadanie nr {nr_zadania+1}:", zadanie)



```
