Stwórz klasę bazową Pojazd z atrybutami: marka , model , rok .

Stwórz klasę potomną Samochod z dodatkowym atrybutem liczba_drzwi .

Dodaj metodę pokaz_info() w obu klasach.

Stwórz obiekt Samochod i wywołaj metodę.

```python
class Pojazd:
    def __init__(self, marka: str, model: str, rok: int):
        self.marka = marka
        self.model = model
        self.rok = rok
    
    def pokaz_info(self):
        print(f"Marka: {self.marka}")
        print(f"Model: {self.model}")
        print(f"Rok: {self.rok}")

class Samochod(Pojazd):
    def __init__(self, marka: str, model: str, rok:int, liczba_drzwi: int):
        super().__init__(marka, model, rok)
        self.liczba_drzwi = liczba_drzwi

    def pokaz_info(self):
        super().pokaz_info()
        print(f"Liczba drzwi: {self.liczba_drzwi}")


samochod = Samochod("Toyota", "Yaris Cross", 2025, 5)
samochod.pokaz_info()

```
