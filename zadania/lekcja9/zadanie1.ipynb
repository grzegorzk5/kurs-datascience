Stwórz klasę Produkt z atrybutami: nazwa , cena , ilosc_w_magazynie .

Dodaj metody:
- pokaz_info() – wyświetla informacje o produkcie
- sprzedaj(ilosc) – zmniejsza ilosc_w_magazynie o podaną wartość
- dostawa(ilosc) – zwiększa ilosc_w_magazynie

Stwórz 3 obiekty i przetestuj metody.

```python
class Produkt:
    def __init__(self, nazwa: str, cena: float, ilosc_w_magazynie: int):
        self.nazwa = nazwa
        self.cena = cena
        self.ilosc_w_magazynie = ilosc_w_magazynie

    def sprzedaj(self, ilosc:int):
        self.ilosc_w_magazynie -= ilosc

    def dostawa(self, ilosc:int):
        self.ilosc_w_magazynie += ilosc

    def pokaz_info(self):
        print(f"Nazwa: {self.nazwa}, cena: {self.cena} zł, Ilość w magazynie: {self.ilosc_w_magazynie} szt.")

produkt1 = Produkt("Zeszyt A4 96 kartek", 10.95, 40)
produkt1.dostawa(4)
produkt1.sprzedaj(25)
produkt1.pokaz_info()

produkt2 = Produkt("Długopis", 0.85, 82)
produkt2.dostawa(28)
produkt2.sprzedaj(40)
produkt2.pokaz_info()

produkt3 = Produkt("Zestaw geometryczny", 4.99, 15)
produkt3.dostawa(5)
produkt3.sprzedaj(6)
produkt3.pokaz_info()
```
