Stwórz klasę BankAccount z atrybutami: wlasciciel , saldo .

Metody:
- wplac(kwota) – zwiększa saldo
- wyplac(kwota) – zmniejsza saldo (tylko jeśli wystarczy środków)
- pokaz_saldo() – wyświetla saldo

Stwórz kilka kont i przetestuj operacje.

```python
class BankAccount:
    def __init__(self, wlasciciel:str, saldo:float):
        self.wlasciciel = wlasciciel
        self.saldo = saldo
    
    def wplac(self, kwota:float):
        self.saldo += kwota

    def wyplac(self, kwota:float) -> bool:
        if self.saldo < kwota:
            return False
        
        self.saldo -= kwota
        return True
    
    def pokaz_saldo(self):
        print(f"Właściciel konta: {self.wlasciciel}, Saldo: {self.saldo}")

konto1 = BankAccount('Ryszard Majewski', 100000)
konto1.pokaz_saldo()
print("Wpłata 5400 zł.")
konto1.wplac(5400)
konto1.pokaz_saldo()

print("Wypłata 1000 zł")
if konto1.wyplac(1000):
    print("Wypłacono.")
else:
    print("Nie wystarczająca ilość środków na koncie")

konto1.pokaz_saldo()

print()

konto2 = BankAccount('Anna Malinowska', 650000)
konto2.pokaz_saldo()

print("Wypłata 2500 zł")
if konto2.wyplac(2500):
    print("Wypłacono.")
else:
    print("Nie wystarczająca ilość środków na koncie")

konto2.pokaz_saldo()

print()
konto3 = BankAccount('Jan Kowalski', 500)

print("Wypłata 900 zł")

if konto3.wyplac(900):
    print("Wypłacono.")
else:
    print("Nie wystarczająca ilość środków na koncie")

konto3.pokaz_saldo()

```
