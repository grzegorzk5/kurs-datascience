Masz dwie tablice reprezentujące ceny produktów przed i po rabacie:

before = np.array([100, 250, 75, 320, 180, 450, 90])
after = np.array([80, 200, 60, 280, 150, 400, 75])

Oblicz i wypisz:
- Różnicę między cenami (oszczędności)
- Procent rabatu dla każdego produktu
- Średnią oszczędność

Wymagania:
- Użyj operacji wektorowych (bez pętli)
- Wynik zaokrąglij do 2 miejsc po przecinku

```python
import numpy as np

before = np.array([100, 250, 75, 320, 180, 450, 90])
after = np.array([80, 200, 60, 280, 150, 400, 75])

oszczednosci = before - after
print("Różnice między cenami (oszczędności): ", oszczednosci)

rabaty_procentowe = np.round((oszczednosci / before) * 100, 2)
print("Procent rabatu dla każdego produktu: ", rabaty_procentowe)

srednia_oszczednosc = np.mean(oszczednosci)
print("Średnia oszczędność: ", np.round(srednia_oszczednosc,2))
```
