Utwórz tablicę NumPy zawierającą liczby od 1 do 100. Oblicz i wypisz:
- Sumę wszystkich elementów
- Średnią arytmetyczną
- Medianę
- Wartość maksymalną i minimalną

Wymagania:
- Użyj np.arange() do stworzenia tablicy
- Użyj funkcji agregujących NumPy ( np.sum , np.mean , etc.)

```python
import numpy as np

tablica = np.arange(1, 101)

print("Suma wszystkich elementów: ", np.sum(tablica))
print("Średnia arytmetyczna: ", np.mean(tablica))
print("Mediana: ", np.median(tablica))
print("Wartość minimalna: ", np.min(tablica))
print("Wartość maksymalna: ", np.max(tablica))
```
