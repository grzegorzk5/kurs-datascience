Dana tablica reprezentująca sprzedaż 4 produktów w 7 dni:

sales = np.random.randint(10, 100, size=(4, 7))

Oblicz i wypisz:
- Średnią sprzedaż per produkt (axis=1)
- Średnią sprzedaż per dzień (axis=0)
- Dzień z najwyższą sprzedażą (np.argmax)
- Produkt z najniższą średnią sprzedażą

```python
import numpy as np

sales = np.random.randint(10, 100, size=(4, 7))

print("Sprzedaż 4 produktów w 7 dni:\n")
print(sales)
print()

print("\n--- Średnia sprzedaż per produkt ---")
produkt_means = np.mean(sales, axis=1)
for nr, srednia in enumerate(produkt_means):
    print(f"Produkt {nr+1}: {srednia:.1f}")

print("\n--- Średnia sprzedaż per dzień ---")
dzien_means = np.mean(sales, axis=0)
for dzien, srednia in enumerate(dzien_means):
    print(f"Dzień {dzien+1}: {srednia:.1f}")

dzien_najwyzsza_sprzedaza_indeks = np.argmax(dzien_means)
print(f"\nDzień z najwyższą sprzedażą:")
print(f"Dzień {dzien_najwyzsza_sprzedaza_indeks + 1} (średnia: {dzien_means[dzien_najwyzsza_sprzedaza_indeks]:.1f})")

produkt_najnizsza_srednia_sprzedaz_indeks = np.argmin(produkt_means)
print(f"\nProdukt z najniższą średnią sprzedażą:")
print(f"Produkt {produkt_najnizsza_srednia_sprzedaz_indeks + 1} (średnia: {produkt_means[produkt_najnizsza_srednia_sprzedaz_indeks]:.1f})")
```
