Masz tablicę z oceną studentów:

grades = np.array([78, 45, 92, 67, 34, 88, 56, 91, 23, 85, 73, 95])

Wykonaj:
- Znajdź wszystkie oceny >= 80 (oceny wysokie)
- Znajdź wszystkie oceny < 50 (oceny niedostateczne)
- Policz ile jest ocen w przedziale [60, 80)
- Zmień wszystkie oceny < 50 na 50 (podniesienie do minimum)

```python
import numpy as np

grades = np.array([78, 45, 92, 67, 34, 88, 56, 91, 23, 85, 73, 95])

print("Tablica z oceną studentów:")
print(grades)

print("\nWszystkie oceny >= 80 (oceny wysokie):", grades[grades >= 80])
print("Wszystkie oceny < 50 (oceny niedostateczne):", grades[grades < 50])

ilosc_ocen_w_przedziale = np.sum((grades >= 60) & (grades < 80))
print("Ilość ocen w przedziale [60, 80):", ilosc_ocen_w_przedziale)

grades_kopia = grades.copy()
grades_kopia[grades_kopia < 50] = 50
print("Po zmianie wszystkich ocen < 50 na 50:", grades_kopia)
```
