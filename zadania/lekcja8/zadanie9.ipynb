Napisz funkcję normalize_minmax(dane) , która normalizuje listę liczb do zakresu [0, 1] używając metody min-max.

Dataset:
wynagrodzenia = [3500, 4200, 5800, 7200, 9500, 4800, 6100]

Wymagania:
- Wzór: (x - min) / (max - min)
- Funkcja zwraca nową listę znormalizowanych wartości
- Zaokrąglij do 3 miejsc po przecinku

Oczekiwany rezultat:
- Lista wartości w zakresie [0.0, 1.0]

```python
def normalize_minmax(dane):
    if not dane:
        return []

    min_val = min(dane)
    max_val = max(dane)
    zakres = max_val - min_val

    znormalizowane = [round((x - min_val) / zakres, 3) for x in dane]
    return znormalizowane

wynagrodzenia = [3500, 4200, 5800, 7200, 9500, 4800, 6100]

print(f"Lista znormalizowanych wartości: {normalize_minmax(wynagrodzenia)}")
```
