Napisz funkcję powitanie(imie, jezyk="polski") , która wyświetla powitanie w wybranym języku (polski, angielski, niemiecki).

Dataset:
- Brak (funkcja operuje na stringach)

Wymagania:
- Domyślny język to polski
- Obsługa 3 języków
- Wywołaj funkcję 3 razy z różnymi parametrami

Oczekiwany rezultat:
- powitanie("Anna") → "Cześć, Anna!"
- powitanie("Anna", "angielski") → "Hello, Anna!"

```python
def powitanie(imie, jezyk="polski"):
    tekst = ""
    if jezyk == "polski":
        tekst = f"Cześć, {imie}"
    elif jezyk == "angielski":
        tekst = f"Hello, {imie}"
    elif jezyk == "niemiecki":
        tekst = f"Hallo, {imie}"
    return tekst

print(f'{powitanie("Anna")}')
print(f'{powitanie("Anna", "angielski")}')
print(f'{powitanie("Anna", "niemiecki")}')
```
