Używając list comprehension, stwórz listę zawierającą tylko liczby parzyste z zakresu 1-100.

Dataset:
- Zakres: 1-100

Wymagania:
- Użyj list comprehension z warunkiem
- Wypisz pierwsze 10 liczb parzystych

Oczekiwany rezultat:
- Lista: [2, 4, 6, 8, 10, ...]

```python
liczby_parzyste = [liczba for liczba in range(1, 101) if liczba % 2 == 0]

print(f"Liczby parzyste z zakresu 1 - 100: {liczby_parzyste}")
```
