Wyświetl tabliczkę mnożenia dla liczby 7 (od 7×1 do 7×10)  używając pętli for i range() .

Wymagania:
- for i in range(1, 11):
- Oblicz wynik = 7 * i
- Wyświetl: "7 × 1 = 7"

Oczekiwany rezultat:
- 10 linii z wynikami mnożenia

```python
print(f"=== TABLICZKA MNOŻENIA DLA LICZBY 7 ===\n")
for i in range(1, 11):
    wynik = 7 * i
    print(f"7 × {i} = {wynik}")
```
