Masz listę wyników testów: [78, 92, 65, 88, 95, 71, 85] . Znajdź maksymalny wynik używając pętli for (bez funkcji max() ).

Wymagania:
- Inicjuj maksimum = wyniki[0]
- W pętli porównuj: if wynik > maksimum:
- Aktualizuj maksimum gdy znajdziesz większą wartość

Oczekiwany rezultat:
- Maksymalny wynik: 95

```python
wyniki = [78, 92, 65, 88, 95, 71, 85]

maksimum = wyniki[0]

for wynik in wyniki:
    if wynik > maksimum:
        maksimum = wynik

print(f"Maksymalny wynik: {maksimum}") 
```
