Masz listę temperatur ciała: [36.5, 37.8, 36.2, 38.5, 36.9, 39.2, 37.0] . Wyświetl tylko te pomiary, które wskazują gorączkę (temperatura >= 38.0°C).

Wymagania:
- Pętla for po liście
- Instrukcja if temperatura >= 38.0:
- Wyświetl temperaturę z ostrzeżeniem

Oczekiwany rezultat:
- Lista temperatur >= 38.0 z komunikatem "Gorączka"

```python
temperatury_ciala = [36.5, 37.8, 36.2, 38.5, 36.9, 39.2, 37.0]

for temperatura in temperatury_ciala:
    if temperatura >= 38.0:
        print(f"Temperatura ciała: {temperatura} °C - Gorączka")
```
