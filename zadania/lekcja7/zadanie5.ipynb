Oblicz sumę wszystkich liczb od 1 do 50 używając pętli while .

Wymagania:
- Zmienne: suma = 0 , i = 1
- while i <= 50:
- Dodaj i do sumy i zwiększ i += 1
- Sprawdź wynik wzorem: n*(n+1)/2 = 50*51/2 = 1275

Oczekiwany rezultat:
- Suma: 1275


```python
n = 50
suma = 0
i = 1

while i <= n:
    suma += i
    i += 1

print(f"Suma: {suma}")

sprawdzenie_wyniku = n * (n + 1) / 2
print(f"Sprawdzenie poprawności wyniku: {sprawdzenie_wyniku:.0f}")
```
