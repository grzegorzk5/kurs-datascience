Zadanie 11

Masz dane sprzedaży miesięcznej: [1200, 1350, 1290, 1480, 1520, 1510, 1680, 1720, 1700, 1890, 1950, 2100] . Zidentyfikuj miesiące, w których sprzedaż wzrosła w porównaniu do poprzedniego miesiąca.

Dataset:
- Lista 12 wartości (sprzedaż miesięczna w PLN)

Wymagania:
- Pętla for i in range(1, len(sprzedaz)):
- Porównaj sprzedaz[i] z sprzedaz[i-1]
- Oblicz zmianę procentową
- Wyświetl miesiące ze wzrostem

Oczekiwany rezultat:
- Lista miesięcy ze wzrostem i procentowy przyrost

```python
sprzedaz = [1200, 1350, 1290, 1480, 1520, 1510, 1680, 1720, 1700, 1890, 1950, 2100]
miesiace = ['Styczeń', 'Luty', 'Marzec', 'Kwiecień', 'Maj', 'Czerwiec', 'Lipiec', 'Sierpień', 'Wrzesień', 'Październik', 'Listopad', 'Grudzień']

print(f"{'Miesiąc':<15} {'Procentowy wzrost sprzedaży'}")
print('-'*44)
for i in range(1, len(sprzedaz)):
    if sprzedaz[i] > sprzedaz[i-1]:
        wzrost = (sprzedaz[i] - sprzedaz[i-1]) / sprzedaz[i-1]
        print(f"{miesiace[i]:<15} ⬆{wzrost:.2%}")
```
