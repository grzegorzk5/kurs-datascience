Symuluj 1000 eksperymentów rzutu 20 monetami (p=0.5). Porównaj histogram wyników z teoretycznym rozkładem dwumianowym B(20, 0.5).

Dataset:
- Wygeneruj używając np.random.binomial()

Wymagania:
- Symulacja 1000 eksperymentów
- Histogram wyników symulacji
- Teoretyczny PMF rozkładu B(20, 0.5) nałożony na histogram
- Porównanie średniej empirycznej z teoretyczną

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Parametry
n = 20          # liczba monet w jednym eksperymencie
p = 0.5         # prawdopodobieństwo sukcesu (orła)
size = 1000     # liczba eksperymentów (powtórzeń)

# Symulacja danych
data = np.random.binomial(n, p, size)

# Obliczenia statystyczne
empirical_mean = np.mean(data)
theoretical_mean = n * p

print(f"Średnia empiryczna: {empirical_mean:.3f}")
print(f"Średnia teoretyczna: {theoretical_mean:.3f}")

# Przygotowanie danych do wykresu teoretycznego (PMF)
x = np.arange(0, n + 1)
pmf = binom.pmf(x, n, p)

# Wizualizacja
plt.figure(figsize=(10, 6))

# Histogram wyników symulacji
plt.hist(data, bins=np.arange(n + 2) - 0.5, density=True, alpha=0.6, 
         color='blue', edgecolor='black', label='Symulacja (Histogram)')

# Teoretyczny rozkład dwumianowy (PMF)
plt.stem(x, pmf, linefmt='r-', markerfmt='ro', basefmt=" ", label='Teoretyczny B(20, 0.5)')

plt.title(f'Porównanie symulacji ({size} rzutów) z rozkładem teoretycznym')
plt.xlabel('Liczba sukcesów (orłów)')
plt.ylabel('Prawdopodobieństwo')
plt.xticks(x)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```
