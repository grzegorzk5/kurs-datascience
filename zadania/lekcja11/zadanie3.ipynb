Z datasetu Titanic wybierz tylko kolumny: Name, Age, Sex, Survived.

Wymagania:
- Stwórz nowy DataFrame z wybranymi kolumnami
- Wyświetl pierwsze 10 wierszy
- Zapisz do pliku 'titanic_subset.csv'

```python
import pandas as pd

url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
titanic = pd.read_csv(url)

titanic_subset = titanic[['Name', 'Age', 'Sex', 'Survived']]

print("Wybrane kolumny z datasetu Titanic (Name, Age, Sex, Survived):")
print(titanic_subset.head(10))
titanic_subset.to_csv("titanic_subset.csv")
```
