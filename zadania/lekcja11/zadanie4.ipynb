Znajdź wszystkich pasażerów Titanica, którzy:
- Mieli mniej niż 18 lat
- Przeżyli katastrofę

Wymagania:
- Użyj boolean indexing
- Policz ile było takich osób
- Wyświetl ich imiona i wiek

```python
import pandas as pd

url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
titanic = pd.read_csv(url)

ocaleni_niepelnoletni_pasazerowie = titanic[(titanic['Age'] < 18) & (titanic['Survived'] == 1)]

liczba_ocalonych_niepelnoletnich_pasazerow = len(ocaleni_niepelnoletni_pasazerowie)
print(f"\nLiczba ocalonych niepełnoletnich pasażerów: {liczba_ocalonych_niepelnoletnich_pasazerow}")

print("\nNieletni pasażerowie, którzy ocaleli z katastrofy:")
print(ocaleni_niepelnoletni_pasazerowie[['Name','Age']].head())
```
