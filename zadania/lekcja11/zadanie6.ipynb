Wypełnij brakujące wartości w kolumnie 'Age' średnią.

Wymagania:
- Oblicz średnią wieku (bez NaN)
- Wypełnij NaN tą średnią
- Sprawdź czy nie ma już NaN w 'Age'

```python
import pandas as pd

url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
titanic = pd.read_csv(url)

# Obliczanie średniej wieku (bez NaN)
# mean() - automatycznie pomija wartości NaN
srednia_wieku_bez_nan = titanic['Age'].mean()

# Wypełnienie brakujących wartości w kolumnie 'Age' średnią 
titanic_wypelniony_wiek = titanic.copy()
titanic_wypelniony_wiek['Age'] = titanic_wypelniony_wiek['Age'].fillna(srednia_wieku_bez_nan)

# Sprawdzenie czy w kolumnie 'Age' są jeszcze jakieś NaN 
liczba_nan_w_wieku = titanic_wypelniony_wiek['Age'].isnull().sum()
# liczba_nan_w_wieku = len(titanic_wypelniony_wiek[titanic_wypelniony_wiek['Age'].isnull()]) # tak również można obliczyć ilość NaN

print(f"Obliczona średnia wieku (bez NaN): {srednia_wieku_bez_nan}")
print(f"Liczba NaN w kolumnie 'Age': {liczba_nan_w_wieku}")

```
