Posortuj pasażerów Titanica po wieku (od najmłodszego do najstarszego).

Wymagania:
- Użyj .sort_values()
- Wyświetl 10 najmłodszych
- Wyświetl 10 najstarszych

```python
# I sposób rozwiązania zadania
import pandas as pd

url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
titanic = pd.read_csv(url)

# Sortownie pasażerów Titanica po wieku od najmłodzego do najstarszego. Pasażerowie bez podanego wieku (wartości NaN) będą na końcu
pasazerowie_posortowani_rosnaco_po_wieku = titanic.sort_values('Age', ascending=True, na_position='last')

print("\n10 najmłodszych pasażerów:")
print(pasazerowie_posortowani_rosnaco_po_wieku.head(10))

print("\n10 najstarszych pasażerów:")
print(pasazerowie_posortowani_rosnaco_po_wieku[pasazerowie_posortowani_rosnaco_po_wieku['Age'].notnull()].tail(10))
```

```python
# II sposób rozwiązania tego zadania

# Usunięcie wierszy z NaN w kolumnie Age
titanic_pominiete_nan = titanic.dropna(subset=['Age'])

# Sortowanie po wieku (od najmłodszego do najstarszego)
titanic_posortowany = titanic_pominiete_nan.sort_values(by='Age', ascending=True)

print("\n10 najmłodszych pasażerów:")
print(titanic_posortowany.head(10))

print("\n10 najstarszych pasażerów:")
print(titanic_posortowany.tail(10))
```
