Zoptymalizuj typy danych w Titanicu aby zużywać mniej pamięci.

Wymagania:
- Zmień Pclass, Survived, SibSp, Parch na int8
- Zmień Sex, Embarked na category
- Przed i po: wypisz zużycie pamięci (.memory_usage(deep=True))
- Oblicz % redukcji

```python
import pandas as pd

url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
titanic = pd.read_csv(url)

titanic_memory_usage = titanic.memory_usage(deep=True).sum()
print(f"Zużycie pamięci przed optymalizacją: {titanic_memory_usage / 1024:.2f} KB")

# Optymalizacja typów
titanic_optimized = titanic.copy()

titanic_optimized['Pclass'] = titanic_optimized['Pclass'].astype('int8')
titanic_optimized['Survived'] = titanic_optimized['Survived'].astype('int8')
titanic_optimized['SibSp'] = titanic_optimized['SibSp'].astype('int8')
titanic_optimized['Parch'] = titanic_optimized['Parch'].astype('int8')

titanic_optimized['Sex'] = titanic_optimized['Sex'].astype('category')
titanic_optimized['Embarked'] = titanic_optimized['Embarked'].astype('category')

optimized_titanic_memory_usage = titanic_optimized.memory_usage(deep=True).sum()
print(f"Zużycie pamięci po optymalizacji: {optimized_titanic_memory_usage / 1024:.2f} KB")

memory_reduction = (1 - optimized_titanic_memory_usage / titanic_memory_usage) * 100
print(f"Redukcja pamięci: {memory_reduction:.1f}%")
```
