Wczytaj dataset Titanic w batchach po 100 wierszy i oblicz średni wiek w każdym batchu.

Wymagania:
- Użyj chunksize=100
- Dla każdego batcha wypisz numer i średni wiek
- Oblicz globalną średnią ze wszystkich batchy

```python
import pandas as pd

url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'

batch_size = 100
chunks = []
total_rows = 0
age_sum = 0
age_count = 0

for i, chunk in enumerate(pd.read_csv(url, chunksize=batch_size)):
    print(f"Średni wiek w batch {i+1}: {chunk['Age'].mean():.2f}")
    total_rows += len(chunk)
    age_sum += chunk['Age'].sum()
    age_count += chunk['Age'].count()

print(f"\nPrzetworzono {total_rows} wierszy")
print(f"Średni wiek ze wszystkich batchy: {(age_sum / age_count):.2f}")
```
