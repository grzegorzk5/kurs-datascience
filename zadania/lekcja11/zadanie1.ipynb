Stwórz DataFrame zawierający informacje o 10 produktach w sklepie: nazwa, cena, ilość na stanie, kategoria.

Wymagania:
- Użyj słownika do utworzenia DataFrame
- Wyświetl pierwsze 5 wierszy
- Oblicz całkowitą wartość towaru (cena × ilość) i dodaj jako nową kolumnę

Oczekiwany rezultat:
- DataFrame z 10 produktami i kolumną 'WartośćCałkowita'

```python
import pandas as pd

data = {
    "Nazwa": ["Czajnik elektryczny", "Mysz bezprzewodowa", "Zeszyt A4 96 kartek", "Powerbank 10000mAh", "Monitor 24 cale", "Pendrive 64GB", "Kabel HDMI 2m", "Waga kuchenna", "Odkurzacz pionowy", "Długopis"],
    "Cena": [129.00, 55.00, 10.50, 79.90, 599.00, 35.00, 29.00, 45.00, 1899.00, 0.90],
    "Ilosc_na_stanie": [24, 28, 141, 22, 7, 50, 40, 30, 15, 219],
    "Kategoria": ["AGD", "Elektronika", "Artykuły biurowe", "Elektronika", "Elektronika", "Elektronika", "Elektronika", "AGD", "AGD", "Artykuły biurowe"]
}

df = pd.DataFrame(data)
print("\nDataFrame - Produkty")
print(df.head())

df['WartoscCalkowita'] = df['Cena'] * df["Ilosc_na_stanie"]
print("\nPo dodaniu kolumny WartoscCalkowita:")
print(df.head())
```
