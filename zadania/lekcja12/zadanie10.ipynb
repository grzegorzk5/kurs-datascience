Napisz własną funkcję, która oblicza zakres międzykwartylowy (IQR = Q3 - Q1) i zastosuj ją do grupy.

Dataset: California Housing

Wymagania:
- Napisz funkcję iqr_func(x) obliczającą IQR
- Zastosuj ją do MedHouseVal według regionów
- Porównaj IQR między regionami

Oczekiwany rezultat:
- Tabela z IQR dla każdego regionu

```python
from sklearn.datasets import fetch_california_housing
import pandas as pd

# Wczytanie danych
housing = fetch_california_housing(as_frame=True)
df_housing = housing.frame

def iqr_func(x):
    Q1 = x.quantile(0.25)
    Q3 = x.quantile(0.75)
    IQR = Q3 - Q1
    return IQR

# Tworzenie kategorii geograficznych (regiony) na podstawie Latitude/Longitude
# Północ/Południe: Latitude > 36
# Wschód/Zachód: Longitude > -119
df_housing['Region'] = 'Unknown'
df_housing.loc[(df_housing['Latitude'] >= 36) & (df_housing['Longitude'] >= -119), 'Region'] = 'NE'     # Północny Wschód
df_housing.loc[(df_housing['Latitude'] >= 36) & (df_housing['Longitude'] < -119), 'Region'] = 'NW'      # Północny Zachód
df_housing.loc[(df_housing['Latitude'] < 36) & (df_housing['Longitude'] >= -119), 'Region'] = 'SE'      # Południowy Wschód
df_housing.loc[(df_housing['Latitude'] < 36) & (df_housing['Longitude'] < -119), 'Region'] = 'SW'       # Południowy Zachód
```

```python
# I sposób z użyciem .apply()
# Zastosowanie funkcji iqr_func do grupy MedHouseVal według Regionu
medhouseval_region = df_housing.groupby('Region', group_keys=False)['MedHouseVal'].apply(iqr_func).reset_index()
medhouseval_region.columns = ['Region', 'MedHouseVal_IQR']

print("=== Tabela z IQR dla każdego regionu ===")
print(medhouseval_region)
```

```python
# II sposób z użyciem .agg()
# Zastosowanie funkcji iqr_func do grupy MedHouseVal według Regionu
medhouseval_region = df_housing.groupby('Region')['MedHouseVal'].agg(iqr_func).reset_index()
medhouseval_region.columns = ['Region', 'MedHouseVal_IQR']

print("=== Tabela z IQR dla każdego regionu ===")
print(medhouseval_region)
```
