Dodaj do datasetu Titanic nową kolumnę Avg_Fare_Pclass , która zawiera średnią opłatę ( Fare ) dla każdej klasy.

Dataset: Titanic

Wymagania:
- Użyj .groupby().transform('mean')
- Dodaj kolumnę do oryginalnego DataFrame
- Wyświetl pierwsze 10 wierszy z kolumnami: Pclass , Fare , Avg_Fare_Pclass

Oczekiwany rezultat:
- DataFrame z nową kolumną, każdy wiersz ma średnią swojej klasy

```python
import pandas as pd

url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
titanic = pd.read_csv(url)

titanic['Avg_Fare_Pclass'] = titanic.groupby('Pclass')['Fare'].transform('mean')

print(titanic[['Pclass', 'Fare', 'Avg_Fare_Pclass']].head(10))

```
