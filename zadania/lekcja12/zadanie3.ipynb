Dla datasetu sprzedaży oblicz dla każdego regionu: minimum, maksimum, średnią i odchylenie standardowe sprzedaży.

Wymagania:
- Użyj .agg(['min', 'max', 'mean', 'std'])
- Zaokrąglij wyniki do 2 miejsc po przecinku

```python
import pandas as pd
import numpy as np

np.random.seed(42)
sales = pd.DataFrame({
'Region': np.random.choice(['A', 'B', 'C'], 100),
'Sprzedaż': np.random.randint(50, 500, 100)
})

sales_region_stats = sales.groupby('Region').agg([
    'min',
    'max',
    'mean',
    'std'
]).round(2)

print("Statystyki sprzedaży według regionu")
print(sales_region_stats)
```
