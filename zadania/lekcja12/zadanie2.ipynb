Dla datasetu Titanic policz, ile pasażerów było w każdej klasie i czy przeżyli ( Survived ).

Dataset: Titanic

Wymagania:
- Grupuj według Pclass i Survived
- Użyj .size() lub .count()
- Przedstaw wyniki jako tabelę pivot

Oczekiwany rezultat:
- Tabela pokazująca liczbę pasażerów dla każdej kombinacji klasy i statusu przeżycia

```python
import pandas as pd

url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
titanic = pd.read_csv(url)

# rozwiazanie z użyciem size()
survival_pivot = titanic.groupby('Pclass')['Survived'].agg([
    'size',
    'sum'
])
survival_pivot.columns = ['Passenger_count', 'Survivor_count']
print(survival_pivot)
```

```python
# rozwiazanie z użyciem count()
survival_pivot = titanic.groupby('Pclass')['Survived'].agg([
    'count',
    'sum'
])
survival_pivot.columns = ['Passenger_count', 'Survivor_count']
print(survival_pivot)
```
