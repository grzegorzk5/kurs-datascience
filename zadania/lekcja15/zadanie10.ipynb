Wybierz 10 losowych obserwacji z Iris. Oblicz pełną macierz odległości euklidesowych (10×10) między wszystkimi parami. Wizualizuj jako heatmapę.

Dataset:
- Iris (sklearn.datasets.load_iris)

Wymagania:
- Losowo wybierz 10 obserwacji (użyj np.random.choice na indeksach)
- Użyj euclidean_distances z sklearn lub zagnieżdżone pętle
- Wizualizuj macierz jako heatmapę (seaborn.heatmap lub plt.imshow)
- Diagonala powinna być 0 (odległość od samego siebie)

Oczekiwany rezultat:
- Macierz 10×10
- Heatmapa z annotacjami

```python
from sklearn.datasets import load_iris
import numpy as np
from sklearn.metrics.pairwise import euclidean_distances
import seaborn as sns
import matplotlib.pyplot as plt

# Wczytanie zbioru danych Iris
iris = load_iris()
X = iris.data

# Losowy wybór 10 obserwacji
np.random.seed(42) # ustawiony po to aby przy każdym uruchomieniu otrzymać te same "losowe" wyniki
indexes = np.random.choice(X.shape[0], 10, replace=False)
observations = X[indexes]

print("10 losowych obserwacji: ")
print(observations)
print(X.shape[0])

# Obliczanie pełnej macierzy odległości euklidesowych (10 x 10)
distances_matrix = euclidean_distances(observations)

print("\nMacierz odległości euklidesowych (10x10):")
print(distances_matrix)
print(f"\nKształt: {distances_matrix.shape}")

# Wizualizacja macierzy jako heatmap
plt.figure(figsize=(12, 8))
ax = sns.heatmap(distances_matrix, 
                 annot=True, 
                 fmt=".3f",
                 cmap='viridis', 
                 linewidths=.5, 
                 linecolor='white',
                 xticklabels=indexes,
                 yticklabels=indexes
) 
ax.set(title="Macierz odległości euklidesowych (10 losowych obserwacji Iris)")
ax.set(xlabel="Indeks obserwacji", ylabel="Indeks obserwacji")
plt.show()
```
