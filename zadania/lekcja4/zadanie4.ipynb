<!-- Ustandaryzuj nazwę kolumny z datasetu.
Dane:
Nazwa raw: " PATIENT_age_YEARS "
Wymagania:
Usuń białe znaki
Zamień na małe litery
Zastąp _ przez " "
Użyj title case
Oczekiwany rezultat:
Patient Age Years -->

```python
nazwa_kolumny_przed_zmiana = " PATIENT_age_YEARS "
nazwa_kolumny_po_zmianie = nazwa_kolumny_przed_zmiana.strip()
print(f"Etap 1: Nazwa kolumny po usunięciu białych znaków: {nazwa_kolumny_po_zmianie}")
nazwa_kolumny_po_zmianie = nazwa_kolumny_po_zmianie.lower()
print(f"Etap 2: Nazwa kolumny po zamianie na małe litery: {nazwa_kolumny_po_zmianie}")
nazwa_kolumny_po_zmianie = nazwa_kolumny_po_zmianie.replace("_", " ")
print(f"Etap 3: Nazwa kolumny po zastąpieniu _ przez \" \": {nazwa_kolumny_po_zmianie}")
nazwa_kolumny_po_zmianie = nazwa_kolumny_po_zmianie.title()
print(f"Etap 4: Nazwa kolumny po zamianie pierwszej litery w każdym słowie na dużą literę -  Oczekiwany rezultat: {nazwa_kolumny_po_zmianie}")
```
