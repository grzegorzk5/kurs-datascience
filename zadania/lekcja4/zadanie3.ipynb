<!-- Oblicz rozmiary zbiorów treningowego, walidacyjnego i testowego.
Dane:
Całkowita liczba próbek: 1000
Proporcje: 70% trening, 15% walidacja, 15% test
Wymagania:
Użyj mnożenia i dzielenia całkowitego
Upewnij się, że suma = 1000 (obsłuż resztę z dzielenia)
Wyświetl rozmiary wszystkich zbiorów -->

```python
liczba_probek = 1000
trening = liczba_probek * 0.7
walidacja = liczba_probek * 0.15
test = liczba_probek * 0.15

print(f"Rozmiar zbioru trenningowego: {trening}")
print(f"Rozmiar zbioru walidacyjnego: {walidacja}")
print(f"Rozmiar zbioru testowego: {test}")

pelne_batche = liczba_probek // 32
pozostale_probki = liczba_probek % 32

print(f"Rozmiar batch: {pelne_batche}")
print(f"Pozostałe próbki: {pozostale_probki}")
```
