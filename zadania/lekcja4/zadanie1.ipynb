<!-- Napisz program, który obliczy wskaźnik BMI (Body Mass Index).

Dane:
Waga: 70 kg
Wzrost: 1.75 m

Wymagania:
Stwórz zmienne waga i wzrost
Oblicz BMI według wzoru: BMI = waga / (wzrost2)
Wyświetl wynik z 2 miejscami po przecinku
Dodaj interpretację: BMI < 18.5 (niedowaga), 18.5-24.9 (norma), ≥25 (nadwaga)

Oczekiwany rezultat:
BMI: 22.86
Interpretacja: Norma -->

```python
waga= 70
wzrost = 1.75

def oblicz_BMI(waga:float, wzrost:float):
    BMI = waga / wzrost ** 2
    return BMI

def interpretacja_BMI(BMI:float):

    if BMI < 18.5:
        interpretacja="Niedowaga"
    elif BMI <= 24.9:
        interpretacja="Norma"
    else:
        interpretacja="Nadwaga"
    
    return interpretacja

BMI = oblicz_BMI(waga, wzrost)
print(f"BMI: {BMI:.2f}")
print(interpretacja_BMI(BMI))
```
