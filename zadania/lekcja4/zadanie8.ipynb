<!-- Oblicz learning rate dla schedulera.
Dane:
Początkowy learning rate: 0.1
Współczynnik decay: 0.95
Numer epoki: 10
Wymagania:
Użyj wzoru: lr_new = lr_initial × (decay ^ epoch)
Operator potęgowania **
Wyświetl wynik w notacji naukowej ( .2e )
Oczekiwany rezultat:
Learning rate po 10 epokach: 5.99e-02 -->

```python
learning_rate_initial = 0.1
decay_rate = 0.95
epoch_number = 10

learning_rate_new = learning_rate_initial * (decay_rate ** epoch_number)

print(f"Learning rate po 10 epokach: {learning_rate_new:.2e}")
```
