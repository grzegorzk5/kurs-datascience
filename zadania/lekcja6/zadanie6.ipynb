Stwórz krotkę z danymi studenta: (imię, nazwisko, wiek, średnia_ocen) . Użyj tuple unpacking by przypisać każdy element do osobnej zmiennej i wypisz informacje.

Wymagania:
- Stwórz krotkę z przykładowymi danymi
- Użyj unpacking
- Wypisz sformatowany wynik

```python
student = ("Jan", "Kowalski", 21, 5.0)
imie, nazwisko, wiek, srednia_ocen = student

print(f"{imie} {nazwisko}, {wiek} lat, średnia ocen: {srednia_ocen}.")
```
