Masz listę z duplikatami: [1, 2, 3, 2, 4, 1, 5, 3, 6, 4, 7] . Usuń duplikaty używając zbioru i przekonwertuj z powrotem na listę. Wypisz obie listy i porównaj długości.

Wymagania:
- Użyj set() do usunięcia duplikatów
- Wypisz przed i po


```python
list_with_duplicates = [1, 2, 3, 2, 4, 1, 5, 3, 6, 4, 7]

print(f"Lista z duplikatami: {list_with_duplicates}")

list_of_unique_elements = list(set(list_with_duplicates)) # Usunięcie duplikatów i utworzenie listy z unikatowymi wartościami
print(f"Lista bez duplikatów: {list_of_unique_elements}")

print(f"\nDługość listy z duplikatami: {len(list_with_duplicates)}")
print(f"Długość listy bez duplikatów: {len(list_of_unique_elements)}")

print(f"Lista z duplikatami jest dłuższa od listy bez duplikatów o {len(list_with_duplicates) - len(list_of_unique_elements)}")
```
