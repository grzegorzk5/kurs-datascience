Masz listę kategorii od użytkowników: ['sport', 'muzyka', 'SPORT', 'film', 'xyz',
'nauka', 'gaming'] . Dozwolone kategorie to {'sport', 'muzyka', 'film', 'nauka',
'gry'} . Znajdź nieprawidłowe kategorie (uwzględnij wielkość liter - konwertuj na małe).

Wymagania:
- Konwertuj wszystkie kategorie na małe litery
- Użyj operacji zbiorowych by znaleźć nieprawidłowe
- Wypisz prawidłowe i nieprawidłowe osobno

```python
kategorie = ['sport', 'muzyka', 'SPORT', 'film', 'xyz', 'nauka', 'gaming']

dozwolone_kategorie = {'sport', 'muzyka', 'film', 'nauka', 'gry'}

kategorie_znormalizowane = [kategoria.lower() for kategoria in kategorie]

kategorie_znormalizowane_zbior = set(kategorie_znormalizowane)

nieprawidlowe_kategorie = kategorie_znormalizowane_zbior - dozwolone_kategorie
prawidlowe_kategorie = kategorie_znormalizowane_zbior & dozwolone_kategorie

print(f"Prawidłowe kategorie: {prawidlowe_kategorie}")
print(f"Nieprawidłowe kategorie: {nieprawidlowe_kategorie}")

```
