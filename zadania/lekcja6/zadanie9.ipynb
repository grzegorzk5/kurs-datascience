Masz słownik mapujący nazwy produktów na kategorie: {'laptop': 'elektronika',
'klawiatura': 'elektronika', 'krzesło': 'meble', 'biurko': 'meble'} . Stwórz odwrotny słownik gdzie klucze to kategorie, a wartości to listy produktów.

Wymagania:
- Iteruj po oryginalnym słowniku
- Zgrupuj produkty według kategorii
- Wynik: {'elektronika': ['laptop', 'klawiatura'], 'meble': [...]}

Oczekiwany wynik:
- Słownik kategoria: lista_produktów

```python
slownik = {'laptop': 'elektronika',
'klawiatura': 'elektronika', 'krzesło': 'meble', 'biurko': 'meble'}

odwrotny_slownik = dict()
for nazwa_produktu, kategoria in slownik.items():
    if kategoria not in odwrotny_slownik:
        odwrotny_slownik[kategoria] = []
    
    odwrotny_slownik[kategoria].append(nazwa_produktu)

print(f"{odwrotny_slownik}")
```
