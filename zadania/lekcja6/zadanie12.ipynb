Masz listę transakcji (krotek): [('2024-01-15', 'Elektronika', 1200), ('2024-01-16', 'Meble', 800), ('2024-01-15', 'Elektronika', 350), ...] . Stwórz słownik gdzie kluczem jest data, a wartością słownik z sumą sprzedaży dla każdej kategorii tego dnia.

Dataset:

transakcje = [
('2024-01-15', 'Elektronika', 1200),
('2024-01-15', 'Meble', 450),
('2024-01-16', 'Elektronika', 800),
('2024-01-16', 'Elektronika', 350),
('2024-01-15', 'Odzież', 200),
('2024-01-16', 'Meble', 600)
]

Wymagania:
- Zagnieżdżone słowniki: {data: {kategoria: suma}}
- Iteruj po transakcjach i akumuluj sumy

Oczekiwany wynik:
- Słownik typu: {'2024-01-15': {'Elektronika': 1200, 'Meble': 450, ...}, ...}

```python
transakcje = [
    ('2024-01-15', 'Elektronika', 1200),
    ('2024-01-15', 'Meble', 450),
    ('2024-01-16', 'Elektronika', 800),
    ('2024-01-16', 'Elektronika', 350),
    ('2024-01-15', 'Odzież', 200),
    ('2024-01-16', 'Meble', 600)
]

slownik_wynikowy = dict()
for data_sprzedazy, kategoria, suma_sprzedazy in transakcje:
    if data_sprzedazy not in slownik_wynikowy:
        slownik_wynikowy[data_sprzedazy] = {kategoria: suma_sprzedazy}
    else:
        slownik_wynikowy[data_sprzedazy][kategoria] = slownik_wynikowy[data_sprzedazy].get(kategoria, 0) + suma_sprzedazy
 
print(slownik_wynikowy)
```
