Masz dwa zbiory: A = {1, 2, 3, 4, 5} i B = {4, 5, 6, 7, 8} . Oblicz i wypisz:
- Sumę zbiorów
- Przecięcie
- Różnicę A - B
- Różnicę symetryczną

Wymagania:
- Użyj operatorów zbiorowych ( | , & , - , ^ )
- Wypisz każdy wynik z opisem

```python
A = {1, 2, 3, 4, 5}
B = {4, 5, 6, 7, 8}

union_of_sets_A_and_B = A | B
print(f"Suma zbiorów: {union_of_sets_A_and_B}")

common_elements_in_A_and_B = A & B
print(f"Wspólne elementy w zbiorze A i B: {common_elements_in_A_and_B}")

elements_in_set_A_but_not_in_B = A - B
print(f"Elementy należące tylko do zbioru A: {elements_in_set_A_but_not_in_B}")

elements_in_A_or_B_but_not_both = A ^ B
print(f"Różnica symetryczna zbiorów A i B: {elements_in_A_or_B_but_not_both}")
```
