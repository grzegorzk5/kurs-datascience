Masz słownik z produktami i cenami: {'jabłko': 3.50, 'banan': 2.20, 'pomarańcza':
4.00} . Dodaj nowy produkt 'gruszka': 3.80 , usuń 'banan' , zmień cenę jabłka na
3.20 i wypisz zaktualizowany słownik.

Wymagania:
- Użyj odpowiednich operacji na słowniku
- Wypisz słownik przed i po zmianach

```python
products_and_prices = {'jabłko': 3.50, 'banan': 2.20, 'pomarańcza': 4.00}

print(f"Słownik przed zmianami: {products_and_prices}")

# Dodanie nowego produktu - gruszka
products_and_prices['gruszka'] = 3.80

# Usunięcie produktu 'banan'
del products_and_prices['banan']

# Zmiana ceny jabłka
products_and_prices['jabłko'] = 3.20

print(f"Słownik po zmianach: {products_and_prices}")
```
