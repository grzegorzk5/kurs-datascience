Stwórz słownik używając dictionary comprehension, gdzie klucze to liczby od 1 do 20, a wartości to ich kwadraty. Następnie przefiltruj ten słownik (też comprehension!) by zawierał tylko liczby parzyste.

Wymagania:
- Użyj dictionary comprehension 2 razy
- Pierwszy słownik: wszystkie liczby 1-20
- Drugi słownik: tylko parzyste klucze

Oczekiwany wynik:
- Dwa słowniki wypisane

```python
kwadraty = dict({liczba:liczba**2 for liczba in range(1, 21)})
print(f"Słownik z wszystkimi kluczami: {kwadraty}")

parzyste_kwadraty = {liczba:kwadrat for liczba, kwadrat in kwadraty.items() if liczba % 2 == 0}
print(f"Słownik z parzystymi kluczami: {parzyste_kwadraty}")
```
