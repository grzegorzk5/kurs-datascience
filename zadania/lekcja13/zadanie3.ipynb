Wymagania:

- Wyświetl 5 najdroższych produktów
- Wyświetl 5 produktów z najmniejszym zapasem
- Wyświetl produkty posortowane alfabetycznie po nazwie

```python
import sqlite3
import pandas as pd

conn = sqlite3.connect("shop.db")

# Wyświetl 5 najdroższych produktów
sql_result = pd.read_sql_query("SELECT name, price FROM products ORDER BY price DESC LIMIT 5", conn)
print("5 najdroższych produktów:")
print(sql_result)

# Wyświetl 5 produktów z najmniejszym zapasem
sql_result = pd.read_sql_query("SELECT name, stock FROM products ORDER BY stock ASC LIMIT 5", conn)
print("\n5 produktów z najmniejszym zapasem")
print(sql_result)

# Wyświetl produkty posortowane alfabetycznie po nazwie
sql_result = pd.read_sql_query("SELECT name FROM products ORDER BY name ASC", conn)
print("\nProdukty posortowane alfabetycznie po nazwie:")
print(sql_result)

conn.close()
```
