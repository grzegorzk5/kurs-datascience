- Używając danych sprzedażowych, pogrupuj według kategorii
- Użyj HAVING aby znaleźć tylko kategorie z przychodem > 1000 PLN
- Oblicz liczbę transakcji, łączny przychód i średnią dla każdej kategorii
- Posortuj według łącznego przychodu

```python
import sqlite3
import pandas as pd

conn = sqlite3.connect(':memory:')

# 1. Klienci
customers = pd.DataFrame({
    'customer_id': [1, 2, 3, 4],
    'name': ['Anna Nowak', 'Jan Kowalski', 'Maria Lewandowska', 'Piotr Wiśniewski'],
    'city': ['Warszawa', 'Kraków', 'Wrocław', 'Poznań']
})

# 2. Zamówienia
orders = pd.DataFrame({
    'order_id': [1, 2, 3, 4, 5],
    'customer_id': [1, 2, 1, 3, 2],
    'order_date': pd.to_datetime(['2024-01-10', '2024-01-12', '2024-01-15', '2024-01-20', '2024-01-25'])
})

# 3. Szczegóły zamówień - produkty, ceny i kategorie
order_items = pd.DataFrame({
    'item_id': [1, 2, 3, 4, 5, 6, 7],
    'order_id': [1, 1, 2, 3, 4, 5, 5],
    'product_name': ['Smartfon', 'Laptop', 'Mysz', 'Klawiatura', 'Monitor', 'Pralka', 'Lodówka'],
    'category': ['Elektronika', 'Elektronika', 'Elektronika', 'Elektronika', 'SElektronika', 'AGD', 'AGD'],
    'price': [2100.0, 3299, 85.0, 150.0, 800.0, 1499.00, 2199.00],
    'quantity': [1, 2, 1, 3, 1, 1, 2]
})

customers.to_sql('customers', conn, index=False)
orders.to_sql('orders', conn, index=False)
order_items.to_sql('order_items', conn, index=False)

"""
- Używając danych sprzedażowych, pogrupuj według kategorii
- Użyj HAVING aby znaleźć tylko kategorie z przychodem > 1000 PLN
- Oblicz liczbę transakcji, łączny przychód i średnią dla każdej kategorii
- Posortuj według łącznego przychodu
"""

query = """
SELECT
    oi.category,
    COUNT(*) AS number_of_transactions,
    SUM(oi.price * oi.quantity) AS total_revenue,
    ROUND(AVG(oi.price * oi.quantity), 2) AS average_value
FROM customers c
INNER JOIN orders o ON c.customer_id = o.customer_id
INNER JOIN order_items oi ON o.order_id = oi.order_id
GROUP BY category
HAVING total_revenue > 1000
ORDER BY total_revenue DESC
"""

sql_result = pd.read_sql_query(query, conn)
print("Wyniki sprzedaży: ")
print(sql_result)
```
