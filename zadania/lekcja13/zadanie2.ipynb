Używając tabeli products z zadania 1:

Wymagania:
- Znajdź produkty droższe niż 100 PLN
- Znajdź produkty z kategorii "Elektronika" LUB z ceną < 50 PLN
- Znajdź produkty z zapasem (stock) między 10 a 50

```python
import sqlite3
import pandas as pd

conn = sqlite3.connect("shop.db")

# Produkty droższe niż 100 PLN
sql_result = pd.read_sql_query("SELECT name, category, price, stock FROM products WHERE price > 100", conn)
print("Produkty droższe niż 100 PLN:")
print(sql_result)

# Produkty z kategorii "Elektronika" LUB z ceną < 50 PLN
sql_result = pd.read_sql_query("SELECT name, category, price, stock FROM products WHERE category = 'Elektronika' OR price < 50", conn)
print("\nProdukty z kategorii \"Elektronika\" LUB z ceną < 50 PLN:")
print(sql_result)

# Produkty z zapasem (stock) między 10 a 50
sql_result = pd.read_sql_query("SELECT name, category, price, stock FROM products WHERE stock BETWEEN 10 AND 50", conn)
print("\nProdukty z zapasem (stock) między 10 a 50:")
print(sql_result)

conn.close()
```
