Utwórz bazę danych SQLite z tabelą products zawierającą kolumny: product_id , name, category , price , stock . Wstaw 10 przykładowych produktów.

Wymagania:
- Napisz zapytanie SELECT wybierające wszystkie produkty
- Napisz zapytanie wybierające tylko name i price
- Wyświetl wyniki za pomocą pd.read_sql_query()

```python
import sqlite3
import pandas as pd

# Tworzenie bazy danych w pamięci
# conn = sqlite3.connect(":memory:")
conn = sqlite3.connect("shop.db")
cursor = conn.cursor()

# Tworzenie tabeli products
cursor.execute("""
CREATE TABLE IF NOT EXISTS products (
    product_id INTEGER PRIMARY KEY,
    name TEXT NOT NULL,
    category TEXT,
    price FLOAT,
    stock INTEGER
)
""")

# Wstawianie przykładowych produktów
products_data = [
    (1, "Powerbank 20000 mAh", "Elektronika", 69.99, 90),
    (2, "Pendrive 128GB", "Elektronika", 49.99, 81),
    (3, "Router Wi-Fi 7", "Elektronika", 420.00, 25),
    (4, "Suszarka do włosów 2200W", "Elektronika", 49.50, 12),
    (5, "Pralka", "AGD", 1499.00, 10),
    (6, "Lodówka", "AGD", 2199.00, 7),
    (7, "Zmywarka", "AGD", 1350.00, 12),
    (8, "Papier ksero A4 80g ryza 500 arkuszy", "Artykuły biurowe", 12.71, 53),
    (9, "Teczka z gumką", "Artykuły biurowe", 2.99, 29),
    (10, "Długopis Zestaw 10 szt Niebieski Automatyczny", "Artykuły biurowe", 5.39, 78)
]

cursor.executemany("""
INSERT OR REPLACE INTO products (product_id, name, category, price, stock)
VALUES (?, ?, ?, ?, ?)
""", products_data)

# Zapis zmian
conn.commit()

# Wszystkie produkty
sql_result = pd.read_sql_query("SELECT * FROM products", conn)
print("Wszystkie produkty:")
print(sql_result)

# Tylko name i price produktów
sql_result = pd.read_sql_query("SELECT name, price FROM products", conn)
print("\nNazwa i cena produktów:")
print(sql_result)

conn.close()
```
